```python
def plot_wordcloud(list_words):
    wordcloud = WordCloud(
        width = 2000, 
        height = 1500, 
        random_state=1, 
        background_color='black', 
        margin=20, 
        colormap='Pastel1', 
        collocations=False
    )
    frequencies = pd.Series(list_words).value_counts()
    plt.figure(figsize=(40, 30))
    plt.imshow(wordcloud.generate_from_frequencies(frequencies))
    plt.axis("off");
```

```python
class Tokenizer:
    def __init__(self, pattern='\W', norm=False):
        self._pattern = re.compile(pattern)
        self._norm = norm
        self._morph = pymorphy2.MorphAnalyzer() if norm else None

    def __call__(self, text):
        tokens = [token for token in self._pattern.split(text.lower()) if token != '']
        if self._norm:
            tokens = [self._morph.parse(token)[0].normal_form for token in tokens]
        return tokens
    
tokenizer = Tokenizer(norm=False)
tokenizer_norm = Tokenizer(norm=True)
```

```python
df_vacancy = pd.read_pickle(f'{DATA_PROCESSED}/df_vacancy.pickle')
df_resume = pd.read_pickle(f'{DATA_PROCESSED}/df_resume.pickle')
```

```python
df_resume = df_resume \
    .explode('experienceItem') \
    .reset_index(drop=True)
df_resume = pd.concat([
    df_resume.drop('experienceItem', axis=1), 
    df_resume['experienceItem'].apply(pd.Series).drop(0, axis=1)
], axis=1)
df_resume['last_fl'] = df_resume['starts'] == df_resume.groupby('uuid')['starts'].transform('max')
```

```python
key_skills = df_resume.query('last_fl')['key_skills'].fillna('').parallel_apply(tokenizer_norm)
```

```python
plot_wordcloud(key_skills.explode())
```
